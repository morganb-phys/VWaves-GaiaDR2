# Summary Plot

First, we import the saved data from the text files 'bin_vel.txt' and 'asym_bin.txt', then plot both the asymmetry and the vertical velocity from Gaia to reproduce Figure 7 in the paper.

```python
zbin1,vave,vave_sig= np.loadtxt('bin_vel.txt')
zbin2,A,A_sig= np.loadtxt('asym_bin.txt')
mask= tuple([(zbin2>-1)*(zbin2<1)])
```

```python
bovy_plot.bovy_print(axes_labelsize=17.,text_fontsize=12.,xtick_labelsize=15.,ytick_labelsize=15.)
figsize(7,8)

fig,(ax1,ax2)= subplots(2,sharex=True)
fig.subplots_adjust(hspace=0.05)

ax1.plot(zbin2[mask],A[mask],'k.-')
ax1.plot(-zbin2[mask],-A[mask], 'k.-')
ax1.errorbar(zbin2[mask],A[mask],yerr= A_sig[mask],marker='None',ls='None',c='C0')
ax1.errorbar(-array(zbin2[mask]),-A[mask], yerr= A_sig[mask],marker='None',ls='None',c='C0')
ax1.axhline(0.,ls=':')

ax2.plot(zbin1,vave,'k.-')
ax2.errorbar(zbin1,vave,yerr=vave_sig,marker='None',ls='None')
ax2.set_xlim([-1,1])
ax2.set_ylim([-4,4])

ax1.set_ylabel(r'$A$')
ax2.set_xlabel(r'$z\,(\mathrm{kpc})$')
ax2.set_ylabel(r'$W\,(\mathrm{km\,s}^{-1})$')

savefig('summary_plot.pdf',bbox_inches='tight')
```
